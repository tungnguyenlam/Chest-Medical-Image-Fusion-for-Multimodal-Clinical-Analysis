```python
import pydicom
import matplotlib.pyplot as plt
import os

# Example path from your data directory
# dicom_path = "data/Lung-PET-CT-Dx/Lung_Dx-A0033/CT-3-81295/16.dcm"

def read_and_display_dicom(path):
    if not os.path.exists(path):
        print(f"File not found: {path}")
        return

    # Read the DICOM file
    ds = pydicom.dcmread(path)

    # Print relevant metadata
    print(f"Modality: {ds.Modality}")
    print(f"Patient ID: {ds.PatientID}")
    print(f"Study Description: {getattr(ds, 'StudyDescription', 'N/A')}")
    print(f"Pixel Data Shape: {ds.pixel_array.shape}")

    # Display the image
    plt.figure(figsize=(8, 8))
    plt.imshow(ds.pixel_array, cmap=plt.cm.bone)
    plt.title(f"DICOM Image - {ds.Modality}")
    plt.axis('off')
    plt.show()
```

```python
read_and_display_dicom(ex_dimcom)
```
